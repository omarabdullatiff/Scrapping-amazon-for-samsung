```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

```

```python
# Set up the Chrome WebDriver
service  = Service('/Users/Omar/Documents/Python_projects/webscrapping/chromedriver')
driver = webdriver.Chrome(service=service)
```

```python
# Navigate to the Amazon Egypt search results page for Samsung products
driver.get("https://www.amazon.eg/s?k=samsung&crid=2DX1G747FCN4M&sprefix=sam%2Caps%2C268&ref=nb_sb_ss_mvt-t11-ranker_2_3")
```

```python
# Extract product descriptions
titles = driver.find_elements("xpath", "//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']")
descriptions = []
for title in titles:
    descriptions.append(title.text)
print(descriptions)
```

```python
# Extract product prices
prices = driver.find_elements(By.XPATH, "//span[@class='a-price-whole']")
prices_ls = []
for price in prices:
    prices_ls.append(price.text)
```

```python
# Extract product image URLs
img_urls = driver.find_elements(By.XPATH,"//img[@class='s-image']")
img_links = []
for img in img_urls:
   img_links.append(img.get_attribute('src'))
print(img_links)
```

```python
# Save the extracted data to a CSV file
import csv 

with open('samsung_products.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Description', 'Price', 'Image URL'])
    for desc, price, img in zip(descriptions, prices_ls, img_links):
        writer.writerow([desc, price, img])
```
